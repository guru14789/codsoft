```python
import pandas as pd  # for data manipulation and analysis.
import numpy as np #numerical calculation in Python.
import seaborn as sns #data visualization
import matplotlib.pyplot as plt # data visualization
from sklearn.feature_extraction.text import TfidfVectorizer #converts a collection of raw text documents into a matrix of TF-IDF features.
from sklearn.model_selection import train_test_split #split dataset into training and testing sets
from sklearn.preprocessing import LabelEncoder #converts categorical labels into numeric values
from sklearn.metrics import accuracy_score, classification_report #model evaluation
from sklearn.svm import LinearSVC #for classication problems
```

```python
train_data = pd.read_csv("train_data.txt",sep=':::', names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION']) #in this step we loading the data
train_data.head()# display first 5 lines of train dataset
```

```python
print(train_data.shape) # no of column and rows
```

```python
test_solution_data = pd.read_csv("test_data_solution.txt",sep=':::', names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'])
print(display(test_solution_data.head()))# display first 5 lines of test dataset
```

```python
print(test_solution_data.shape)# no of column and rows
```

```python
test_data = pd.read_csv("test_data.txt",sep=':::', names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'])
print(display(test_data.head()))
```

```python
print(test_data.shape)# no of column and rows
```

```python
train_data['DESCRIPTION'].fillna("", inplace=True) #this step is like handling missing values
test_data['DESCRIPTION'].fillna("", inplace=True)

t_v = TfidfVectorizer(stop_words='english', max_features=100000) #TF-IDF Vectorization:
X_train = t_v.fit_transform(train_data['DESCRIPTION'])
X_test = t_v.transform(test_data['DESCRIPTION'])

label_encoder = LabelEncoder() #This converts the categorical labels into numerical labels so that the machine learning model can work.
y_train = label_encoder.fit_transform(train_data['GENRE'])
y_test = label_encoder.transform(test_solution_data['GENRE'])
```

```python

X_train_sub, X_val, y_train_sub, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42) #here i am split training into two subset one is 20% of the original training data as the validation set to assess the model's performance.

clf = LinearSVC() #The LinearSVC classifier  is being trained on the training subset.
clf.fit(X_train_sub, y_train_sub)

y_val_pred = clf.predict(X_val) #making predictions
print("Validation Accuracy:", accuracy_score(y_val, y_val_pred)) # accuracy of the model
print("Validation Classification Report:\n", classification_report(y_val, y_val_pred))
```

```python
y_pred = clf.predict(X_test) #making predictions
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print("Test Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB
Mnb_classifier = MultinomialNB() # used for binary or multi-class classification problems.
Mnb_classifier.fit(X_train, y_train) #fitting x_train and y_train
```

```python
def predict_movie(description):
    t_v1 = t_v.transform([description]) #transform the Description:
    pred_label = clf.predict(t_v1) #Predict the Genre
    return label_encoder.inverse_transform(pred_label)[0] #Decode the Predicted Label
```
